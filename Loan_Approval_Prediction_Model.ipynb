

First thing before buiding a machine learning model is we need the dataset. So once we have the dataset we need to do the cleansing and we need to do the analysis of the same dataset.

Once we are donw with this process, we are going to split the dataset into two parts. First part is called training dataset(80%) and another part is called testing dataset(20%).

After the splitting of dataset, we are going to build the Machine Learning model. In this project I am going to create a logistic refression ML model. To predict the loan status based upon the certain inputs from the user.

Once the logistic regression model is created, we will be training the model with the training dataset.

Once the training of the model is completed, we will be evaluating that model performance with the test dataset.

Once we are good with all this process, we will be saving this model and will deploy it as a web application


```python
# pandas, scikit-learn, streamlit, pickle are needed for this ML model.

import pandas as pd
```

```python
data = pd.read_csv("/content/loan_approval_dataset.csv")
```

```python
data
```

```python
data.drop(columns=['loan_id'], inplace=True)
```

```python
data.columns
```

```python
# there is a "space" before every column name. Lets remove that.

data.columns = data.columns.str.strip()
```

```python
data.columns

# now cleansing of the column is completed
```

```python
# lets club all the asset value columns into a single column
data.columns
```

```python
data['Assets'] = data.residential_assets_value + data.commercial_assets_value + data.luxury_assets_value + data.bank_asset_value
```

```python
data.drop(columns = ['residential_assets_value','commercial_assets_value','luxury_assets_value','bank_asset_value'],inplace=True)
```

```python
data
```

```python
# to improve the performance of ML model,
# we are going to convert alphabetic or string values into numeric values.
# we have three string value columns - education, self_employed, loan_status
```

```python
# before that lets check duplicate values in the dataset

data.isnull().sum()

# we have no null values in this dataset
# if we have it, use data.dropna(inplace=True)
```

```python
# now lets check distinct values in "education"

data.education.unique()
```

```python
# here we can see there is extra space in column values.
# lets remove this extra space for the whole column

def clean_data(st):
  st = st.strip()
  return st
```

```python
data.education = data.education.apply(clean_data)
```

```python
data.education.unique()

# now extra spaces are gone.
```

```python
# now lets replace "Graduate" and "Not Graduate" with numerical values

data['education'] = data['education'].replace(['Graduate', 'Not Graduate'],[1,0])
```

```python
data
```

```python
data.self_employed.unique()
```

```python
data.self_employed = data.self_employed.apply(clean_data)
```

```python
data.self_employed.unique()
```

```python
data.self_employed = data.self_employed.replace(["No","Yes"],[0,1])
```

```python
data
```

```python
data.loan_status.unique()
```

```python
data.loan_status = data.loan_status.apply(clean_data)
```

```python
data.loan_status.unique()
```

```python
data.loan_status = data.loan_status.replace(['Approved','Rejected'],[1,0])
```

```python
data
```

```python
# now we have 9 columns

# now lets split dataset into training dataset and test dataset

from sklearn.model_selection import train_test_split

# loan_status is output column. Remaining 8 column will be input columns.
```

```python
input_data = data.drop(columns =['loan_status'])
output_data = data['loan_status']
```

```python
input_data

# this gives 8 columns
```

```python
output_data

# only one column
```

```python
# now we have "input dataset" and "output dataset"
# lets split the dataset into - train test data splits
```

```python
x_train,x_test,y_train,y_test = train_test_split(input_data,output_data,test_size=0.2)

# we have to pass the two arguments inside "train_test_split"
# they are "input_data" and "output_data"
# 20% of data goes to test dataset
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape

# 3415 rows became the training dataset
# 854 rows became the test dataset
```

```python
# now we have splitted our dataset to training dataset and testing dataset.
```

```python
# In the dataset, we can see large number of amounts in crores.
# so there will be outliers
# we have to do the scaling of this data. It improves performance of the ML model.

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.fit_transform(x_test)
```

```python
# now lets create the logistic regression model, so lets import it from scikit-learn

from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()

# model is created with the above command.
```

```python
# now lets fit the model with the train dataset
# we are going to train the dataset.

model.fit(x_train_scaled, y_train)
```

```python
# lets evaluate the model score below

model.score(x_test_scaled,y_test)
```

```python
# our model has 90% accuracy.
```

```python
# lets predict some values

pred_data = pd.DataFrame([['2','1','0','9600000','29900000','12','778','50700000']],columns=['no_of_dependents','education','self_employed','income_annum','loan_amount','loan_term','cibil_score','Assets'])
```

```python
# lets scale this dataset with "scaler.transform"

pred_data = scaler.transform(pred_data)
```

```python
# lets predict the value..

model.predict(pred_data)
```

```python
# I got 1, so loan is approved
# for the first row, loan got approved
```
